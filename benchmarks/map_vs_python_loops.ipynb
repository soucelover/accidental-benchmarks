```python
from typing import Iterable


def test_filter(objs: Iterable[object]) -> list[object]:
    return list(filter(bool, objs))


def test_list_compr(objs: Iterable[object]) -> list[object]:
    return [item for item in objs if item]


run_test("test_filter_1000", "test_filter(objs)", "objs = list(range(1000))")
run_test(
    "test_list_compr_1000",
    "test_list_compr(objs)",
    "objs = list(range(1000))",
)

run_test(
    "test_filter_1000000",
    "test_filter(objs)",
    "objs = list(range(1_000_000))",
)
run_test(
    "test_list_compr_1000000",
    "test_list_compr(objs)",
    "objs = list(range(1_000_000))",
)
```
